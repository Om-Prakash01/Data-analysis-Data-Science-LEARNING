# Matplotlib / Data Visualization :-

# Types of Data :-

```python
# import the library

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt # data visualization library
import seaborn as sns  
```

# 2D line plot :-
    • Bivariate Analysis -> Always works on two cols like -> x axis and y axis
    • Categorical -> Numerical and Numerical -> Numerical
    • Use case - Time series data.

```python
# Plotting a simple function 

price = [48000, 54000, 57000, 49000, 47000, 45000] # Numerical
year = [2015, 2016, 2017, 2018, 2019, 2020]  # Categorical 

# you alway put categorical data in x axis and numerical data in y axis.
plt.plot(year, price)

plt.grid(True)
```

```python
# from a pandas dataframes :-

batsman = pd.read_csv('D:\learning\sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
plt.grid()
```

```python
# plotting multiple plots :-

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.grid()
```

```python
# labels title :-

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex code) and line(width and style) and marker(size)

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F')# any color you can pass with hexcode or with name.
plt.plot(batsman['index'], batsman['RG Sharma'], color='black')

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# linestyle(style for lines) function.
# linewidth(width for lines) increase the size of the line.

# linestyle option -> dashdot , solid , dashed, more 
plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker in parameter / markersize also you can set in parameter -> 2D plot line

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location 
# you can give every line each name -> parameter as -> label

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9, label='Rohit') 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend() # loc parameter for shifting legend
```

```python
# limiting axes 

price = [48000, 54000, 57000, 49000, 47000, 45000, 4500000] # Numerical
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]  # Categorical

plt.plot(year, price)

# when will you do trimming in limits of yaxis and xaxis ? -> when you have some outliers in your data 
# and you want to ignore that outlier for better visualization. 
plt.ylim(0, 75000) # trimming y axis
plt.xlim(2017, 2019) # trimming x axis
```

```python
# grid() function :-

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show() function :-

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.show() # to show the plot in a separate window
```

# Scatter Plot :-

    • bivariate Analysis -> Always works on two cols like -> x axis and y axis
    • Numerical -> Numerical
    • Use case :- Finding Corelation between two numerical variables.

    • To show the relationship between two numerical variables.
    • Each point represents an observation in the dataset.

```python
# plt.scatter() simple function :-

x = np.linspace(-10, 10, 50)

y = 10*x + 3 + np.random.randint(0, 300, 50)
y
```

```python
plt.scatter(x, y)
```

```python
# plt.scatter() on pandas data :-
df = pd.read_csv('D:\learning\\batter.csv')
df = df.head(50)
df
```

```python
plt.scatter(df['avg'], df['strike_rate'])
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# marker size(s) and color in scatter plot :-

plt.scatter(df['avg'], df['strike_rate'],color='yellow', marker='*')
plt.title('Avg and SR analysis of Top 50 Batsman')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
```

```python
# size
tips = sns.load_dataset('tips') # dataset in seaborn.


plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*10)
```

```python
# scatterplot using plt.plot :- (faster but not too many parameter)

plt.plot(tips['total_bill'], tips['tip'], 'o') # without the parameter 'o' this plot is 2D plot
```

# Bar Chart :-
    • Bivariate Analysis / Univariate Analysis 
    • Numerical vs Categorical 
    • Use Case -> Aggregate analysis of groups 

```python
# simple bar chart :- 

children = [10, 20, 24, 14, 19]
color = ['red', 'blue', 'green', 'yellow', 'Pink']

plt.bar(color, children) # Vertical bar chart 
```

```python
# bar chart using data :-
```

```python
#  horizontal bar chart :-
plt.barh(color, children)
```

```python
# color and label :-

df = pd.read_csv('D:\learning\\batsman_season_record.csv')
df
```

```python
np.array(df.shape[0])
```

```python
# Multiple Bar Chart :-

# You can also set the color 
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'], width=0.2)
plt.bar(np.arange(df.shape[0]), df['2016'], width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'], width=0.2) 

# xticks :-

plt.xticks(np.arange(df.shape[0]), df['batsman'])
# plt.show()
```

```python
# a problem :-

children = [10, 20, 24, 14, 19]
color = ['red red red red', 'blue blue blue blue', 'green green green green', 'yellow yellow yellow yellow', 'Pink Pink Pink Pink']

plt.bar(color, children) # Vertical bar chart 
plt.xticks(rotation='vertical') # changing the name rotation.
```

```python
# stacked bar chart :-

plt.bar(df['batsman'], df['2015'], label='2015')
plt.bar(df['batsman'], df['2016'], bottom=df['2015'], label='2016')
plt.bar(df['batsman'], df['2017'], bottom=(df['2016'] + df['2015']), label='2017')

plt.legend()
plt.show()
```

# Pie Chart :-
    • Univariate/Bivariate Analysis
    • Categorical vd Numerical 
    • Use Case -> TO find Contribution on a standard scale 

```python
# Simple data Example :-

data = [23, 45, 100, 20, 49]
subject = ['english', 'Science', 'Math', 'IT', 'Management']

plt.pie(data,labels=subject)
# plt.show()
```

```python
# dataset Example :-
df = pd.read_csv('D:\learning\gayle-175.csv')
df


# for percentage parameter
# explode parameter 

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.1, 0, 0, 0, 0.1, 0]) 
plt.show()
```

```python
# explode shadow :-

plt.pie(df['batsman_runs'], labels=df['batsman'], autopct='%0.1f%%', explode=[0.1, 0, 0, 0, 0.1, 0], shadow=True) 
plt.show()
```
