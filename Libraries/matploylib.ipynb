# 2D line plot :-
    • Bivariate Analysis -> Always works on two cols like -> x axis and y axis
    • Categorical -> Numerical and Numerical -> Numerical
    • Use case - Time series data.

```python
# Plotting a simple function 

price = [48000, 54000, 57000, 49000, 47000, 45000] # Numerical
year = [2015, 2016, 2017, 2018, 2019, 2020]  # Categorical 

# you alway put categorical data in x axis and numerical data in y axis.
plt.plot(year, price)

plt.grid(True)
```

```python
# from a pandas dataframes :-

batsman = pd.read_csv('D:\learning\sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
plt.grid()
```

```python
# plotting multiple plots :-

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.grid()
```

```python
# labels title :-

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# colors(hex code) and line(width and style) and marker(size)

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F')# any color you can pass with hexcode or with name.
plt.plot(batsman['index'], batsman['RG Sharma'], color='black')

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# linestyle(style for lines) function.
# linewidth(width for lines) increase the size of the line.

# linestyle option -> dashdot , solid , dashed, more 
plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker in parameter / markersize also you can set in parameter -> 2D plot line

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# legend -> location 
# you can give every line each name -> parameter as -> label

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9, label='Rohit') 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend() # loc parameter for shifting legend
```

```python
# limiting axes 

price = [48000, 54000, 57000, 49000, 47000, 45000, 4500000] # Numerical
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]  # Categorical

plt.plot(year, price)

# when will you do trimming in limits of yaxis and xaxis ? -> when you have some outliers in your data 
# and you want to ignore that outlier for better visualization. 
plt.ylim(0, 75000) # trimming y axis
plt.xlim(2017, 2019) # trimming x axis
```

```python
# grid() function :-

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.grid()
```

```python
# show() function :-

plt.plot(batsman['index'], batsman['V Kohli'],color='#D9F10F', linestyle='dashed', linewidth=3, marker='o', markersize=8)
plt.plot(batsman['index'], batsman['RG Sharma'], color='black', linestyle='dashdot', linewidth=4, marker='D', markersize=9) 

plt.title('Rohit sharma vs Virat Kohli Career Comparison').set_fontstyle('italic')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.show() # to show the plot in a separate window
```

# Colored Scatter Plot :-
    • To add more information to a scatter plot, we can use color to represent a third variable.
    • This is particularly useful when we want to visualize the relationship between three numerical variables.

```python
iris = pd.read_csv('D:\learning\iris.csv')
iris.sample(5)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# Renaming the name of species 

iris['Species'] = iris['Species'].replace({'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2})

iris.sample(5)
```

```python
# cmap  -> colormap
# alpha -> for opacity of color in scatter plot

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='viridis', alpha=1.0) # for changing the color use cmap paramter .
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar() # to show the color scale on the side of the plot.
```

# Plot Size :-

```python
# plt.figure(figsize(15, 7)) -> pass a tuple in parameter ( width, length)

plt.figure(figsize=(10, 6)) # set the size of plot

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='viridis', alpha=1.0) # for changing the color use cmap paramter .
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')

plt.colorbar()
```

# Annotations :- 
    • Annotations are used to add text or labels to specific points on a plot.
    • They help in highlighting important data points or providing additional context to the visualization.

```python
batters = pd.read_csv('D:\\learning\\batter.csv')
batters
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
sample_df
```

```python
plt.figure(figsize=(18, 10)) # (width, length)

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python
# Example of Annotations :-

x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

plt.scatter(x, y)

# you also provide colors and markers
plt.text(1, 6, 'Point 1')
plt.text(2, 7, 'Point 2')
plt.text(3, 8, 'Point 3')
plt.text(4, 9, 'Point 4')
plt.text(5, 10, 'Point 5', fontdict={'color':'red', 'size':15})
```

# Horizontal and vertical lines :-

```python
plt.figure(figsize=(18, 10)) # (width, length)

plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

plt.axhline(130, color='red')
plt.axhline(140, color='green')
plt.axvline(30, color='red')

for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

# Subplots :-

```python
# A diff way to plot graphs 

batters.head()
```

```python
# one way of plotting scatter plot :-

plt.figure(figsize=(15, 6))

plt.scatter(batters['avg'], batters['strike_rate']) 

plt.title('Something')
plt.xlabel('A')
plt.ylabel('B')

plt.show()
```

```python
# Alternate way of plotting scatter plot :-

fig, ax = plt.subplots(figsize=(15, 6))

ax.scatter(batters['avg'], batters['strike_rate'], color='red', marker='+')
ax.set_title('something')
ax.set_xlabel('A')
ax.set_ylabel('B')

fig.show()
```

```python
# subplot :- 2X1 

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(14, 10), sharex=True)

ax[0].scatter(batters['avg'], batters['strike_rate'], s=75, color='skyblue')
ax[1].scatter(batters['avg'], batters['runs'], s=75)

ax[0].set_title('Avg vs Strike rate')
ax[0].set_ylabel('Strike rate')

ax[1].set_title('Avg vs runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
# subplot :- 2X2

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))

ax[0, 0].scatter(batters['avg'], batters['strike_rate'], s=75, color='skyblue')
ax[0, 1].scatter(batters['avg'], batters['runs'], s=75)
ax[1, 0].hist(batters['avg'])
ax[1, 1].hist(batters['runs'])
```

```python
# other way plot same 2X2 subplot :-

fig = plt.figure()

ax1 = fig.add_subplot(2, 2, 1)
ax1.scatter(batters['avg'], batters['strike_rate'], color='red')

ax2 = fig.add_subplot(2, 2, 2)
ax2.hist(batters['runs'])

ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(batters['avg'])

ax4 = fig.add_subplot(2, 2, 4)
ax4.scatter(batters['avg'], batters['runs'], color='red')
```
