# • Plots under this section :- 
    • Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px  # for dataset Only
```

```python
# Dataset Load

tips = sns.load_dataset('tips')
tips
```

```python
# Scatter Plot -> axis level function

sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex', style='time')
```

```python
# Relplot -> figure level function

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='sex', style='time')
```
